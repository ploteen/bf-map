```python

# 라이브러리 추가
import requests; from urllib.parse import urlparse
import pandas as pd
import geopandas
import math
```

```python
df = pd.read_csv('./data.csv')
```

```python
df=df.loc[:,['address','name']]
```

```python
list_address = list(df.loc[:,'address'])
list_name = list(df.loc[:,'name'])
result_list = list()
```

```python
for idx, address in enumerate(list_address):
    if type(address) == float:
        result_list.append([list_name[idx],None,None,None])
        print(idx)
        continue
    url = "https://dapi.kakao.com/v2/local/search/address.json?&query=" + str(address)
    result = requests.get(urlparse(url).geturl(),
                      headers={"Authorization":"KakaoAK 1caf4414034b6d3055b4689983282626"})
    json_obj = result.json()
    for document in json_obj['documents']:
        val = [list_name[idx], document['address_name'], document['y'], document['x']]
        result_list.append(val)
```

```python
result_list
```

```python
df = pd.DataFrame(result_list, columns = ['title', 'address','lat', 'lon'])
```

```python
df.to_csv('new_data.csv')
```
